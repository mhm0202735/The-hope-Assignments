**correlation coefficient**



*   The correlation coefficient is an equation that is used to determine

      the strength of the relation between two variables


*   The correlation coefficient always lies between -1 to +1
*    -1 represents X and Y are negatively correlated
*   +1 represents X and Y are positively correlated

r = n (âˆ‘xy) â€“ âˆ‘x âˆ‘y / âˆš [n* (âˆ‘x2 â€“ (âˆ‘x)2)] * [n* (âˆ‘y2 â€“ (âˆ‘y)2)]


*   r = correlation coefficient

*   n = number of observations

*   x = 1st variable in the context



*   y = 2nd variable




---

Absolute Value of r Strength of Relationship


r < 0.3  ==>  None or very weak


0.3 < r <0.5   ==>  Weak


0.5 < r < 0.7  ==>   Moderate


r > 0.7  ==>   Strong

```python
# Solution 6.3
def calculate_correlation(DF,first_column,second_column):
  return (DF.size * (DF[first_column] * DF[second_column]).sum() - (DF[first_column].sum() *DF[second_column].sum()))/np.sqrt((DF.size * ((DF[first_column] ** 2)).sum() - (DF[first_column].sum() ** 2)) * (DF.size * ((DF[second_column] ** 2)).sum() - (DF[second_column].sum() ** 2)))
r = calculate_correlation(df,"total_rooms","median_house_value")
print(r)

#r = (df.size * (df["total_rooms"] * df["median_house_value"]).sum() - (df["total_rooms"].sum() *df["median_house_value"].sum()))/np.sqrt((df.size * ((df["total_rooms"] ** 2)).sum() - (df["total_rooms"].sum() ** 2)) * (df.size * ((df["median_house_value"] ** 2)).sum() - (df["median_house_value"].sum() ** 2)))
```
